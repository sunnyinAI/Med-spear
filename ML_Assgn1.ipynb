2. Write a function to generated dependent variable column t. [1]

a) Inputs: Data matrix X, weight vector for each column, bias w0, noise variance

b) Output: Target vector t


---



```python
def generate_target_vector(X, weights, bias, noise_var):
    # calculate dot product of X and weight
    dot_product = np.dot(X, weights)
    # add bias
    dot_product += bias
    # add noise
    noise = np.random.normal(0, noise_var, dot_product.shape)
    t = dot_product + noise
    return t
```

```python
X = generate_data_matrix(200, 5)
weights = np.random.randn(5)
bias = 2
noise_var = 0.1
t = generate_target_vector(X, weights, bias, noise_var)
print(t)
```

4. Write a function to compute the mean square error of two vectors y and t. [1]

```python
def mean_square_error(y, t):
    # calculate the difference between y and t
    diff = y - t
    # calculate the mean square error
    mse = np.mean(diff**2)
    return mse
```

6. Write a function to compute the gradient of MSE with respect to its weight vector. [2]

a) Input: X matrix, t vector, and w vector

b) Output: gradient vector


---



```python
def gradient_MSE(X, t, w):
    # calculate the estimate of the dependent variable
    y = np.dot(X, w)
    # calculate the difference between y and t
    diff = y - t
    # calculate the gradient
    gradient = 2 * np.dot(X.T, diff) / X.shape[0]
    return gradient
```

```python
X = generate_data_matrix(200, 5)
weights = np.random.randn(5)
bias = 2
noise_var = 0.1
t = generate_target_vector(X, weights, bias, noise_var)
gradient = gradient_MSE(X, t, weights)
print(gradient)
```

7. Write a function to compute L2 norm of a vector w passed as a numpy array. Exclude bias w0. [1]


---



```python
def l2_norm(w):
    # exclude bias w0
    w = w[1:]
    # calculate the L2 norm
    l2 = np.sqrt(np.dot(w, w))
    return l2
```

```python
weights = np.random.randn(5)
l2 = l2_norm(weights)
print(l2)
```

```python
def gradient_L2(w):
    # exclude bias w0
    w = w[1:]
    # calculate the gradient
    gradient = w / np.linalg.norm(w)
    # insert a 0 in the beginning of the gradient vector to represent w0
    gradient = np.insert(gradient, 0, 0)
    return gradient
```

```python
weights = np.random.randn(5)
gradient = gradient_L2(weights)
print(gradient)
```

10. Write a function to compute the gradient of L1 norm with respect to the weight vectors. [2]

a) Input: X matrix and w vector
b) Output: gradient vector, where gradient with respect to w0 is 0.


---



```python
def gradient_L1(w):
    # exclude bias w0
    w = w[1:]
    # calculate the gradient
    gradient = np.sign(w)
    # insert a 0 in the beginning of the gradient vector to represent w0
    gradient = np.insert(gradient, 0, 0)
    return gradient

    
weights = np.random.randn(5)
gradient = gradient_L1(weights)
print(gradient)
```

11. Write a function for a single update of weights of linear regression using gradient descent. [2]

a) Input: X, t, w, eta, lambda 2, lambda 1. Note that the weight of MSE will be 1

b) Output: updated weight and updated MSE


---



```python
def update_weights(X, t, w, eta, reg_lambda_2, reg_lambda_1):
    # calculate the gradient of MSE
    gradient_mse = gradient_MSE(X, t, w)
    # calculate the gradient of L2 norm
    gradient_l2 = gradient_L2(w)
    # calculate the gradient of L1 norm
    gradient_l1 = gradient_L1(w)
    # calculate the total gradient
    gradient = gradient_mse + reg_lambda_2 * gradient_l2 + reg_lambda_1 * gradient_l1
    # update the weights
    w = w - eta * gradient
    # calculate the estimate of the dependent variable
    y = np.dot(X, w)
    # calculate the mean square error
    mse = mean_square_error(y, t)
    return w, mse


X = generate_data_matrix(10, 5)
weights = np.random.randn(5)
bias = 2
noise_var = 0.1
t = generate_target_vector(X, weights, bias, noise_var)
w, mse = update_weights(X, t, weights, 0.1, 0.01, 0.01)
print(w)
print(mse)
```
